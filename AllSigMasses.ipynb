# Muon Channel

```python
Z1_M500_df = pd.read_csv("../Wprime_VBF_gWWpZ1_M500.csv")
Z1_M1000_df = pd.read_csv("../Wprime_VBF_gWWpZ1.csv")
Z1_M3000_df = pd.read_csv("../Wprime_VBF_gWWpZ1_M3000.csv")

Z2_M500_df = pd.read_csv("../Wprime_VBF_gWWpZ2_M500.csv")
Z2_M1000_df = pd.read_csv("../Wprime_VBF_gWWpZ2.csv")
Z2_M3000_df = pd.read_csv("../Wprime_VBF_gWWpZ2_M3000.csv")
```

```python
Z1_M500_df_mu = basic_allch_cuts(Z1_M500_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)
Z1_M1000_df_mu = basic_allch_cuts(Z1_M1000_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)
Z1_M3000_df_mu = basic_allch_cuts(Z1_M3000_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)

m_T_Z1_M500, m_jj_Z1_M500, Delta_eta_Z1_M500, Mul_eta_Z1_M500 = muVBFvars(Z1_M500_df_mu)
m_T_Z1_M1000, m_jj_Z1_M1000, Delta_eta_Z1_M1000, Mul_eta_Z1_M1000 = muVBFvars(Z1_M1000_df_mu)
m_T_Z1_M3000, m_jj_Z1_M3000, Delta_eta_Z1_M3000, Mul_eta_Z1_M3000 = muVBFvars(Z1_M3000_df_mu)

Z2_M500_df_mu = basic_allch_cuts(Z2_M500_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)
Z2_M1000_df_mu = basic_allch_cuts(Z2_M1000_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)
Z2_M3000_df_mu = basic_allch_cuts(Z2_M3000_df, n_j = 2, pt_j = 60, n_b = 0, met = 50, n_tau = 0, n_e = 0, n_mu = 1, pt_mu = 25)

m_T_Z2_M500, m_jj_Z2_M500, Delta_eta_Z2_M500, Mul_eta_Z2_M500 = muVBFvars(Z2_M500_df_mu)
m_T_Z2_M1000, m_jj_Z2_M1000, Delta_eta_Z2_M1000, Mul_eta_Z2_M1000 = muVBFvars(Z2_M1000_df_mu)
m_T_Z2_M3000, m_jj_Z2_M3000, Delta_eta_Z2_M3000, Mul_eta_Z2_M3000 = muVBFvars(Z2_M3000_df_mu)
```
